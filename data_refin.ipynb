```python
# Group the DataFrame by 'Customer_id' and iterate over each group
for _, group in demographics.groupby('Customer_id'):
    # Check if the group has more than one occurrence
    if len(group) > 1:
        # Generate a unique suffix for each occurrence after the first one
        suffixes = [f'_{i}' for i in range(1, len(group))]
        # Append suffixes to the 'Customer_id' column values
        for i, suffix in enumerate(suffixes):
            demographics.loc[group.index[i + 1], 'Customer_id'] += suffix

# Check the modified DataFrame
print(demographics)
```
